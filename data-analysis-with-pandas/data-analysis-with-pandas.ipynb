## Unique data

```python
def getUnique(df, name):
    return df[name].unique()
```

```python
#Categories
categories = getUnique(df, 'Subject')
#Actor
actor = getUnique(df, 'Actor')
#Actress
actress = getUnique(df, 'Actress')
#Year
year = getUnique(df, 'Year')
```

## Plot: Number of films by actors (Limit by 10)

```python
df['Actor'].value_counts(ascending=False)[:10].plot.bar(title='Number of films by actors')
plt.xlabel("Actors")
plt.ylabel("Number of films");
```

## Plot: Number of films by actresses (Limit by 10)

```python
df['Actress'].value_counts(ascending=False)[:10].plot.bar(title='Number of films by actress')
plt.xlabel("Actresses")
plt.ylabel("Number of films");
```

## Plot: Number of films by directors (Limit by 10)

```python
df['Director'].value_counts(ascending=False)[:10].plot.bar(title='Number of films by directors')
plt.xlabel("Directors")
plt.ylabel("Number of films");
```

## Plot: Number of films by Subject (Category)

```python
df['Subject'].value_counts(ascending=True).plot.barh(title='Number of films by Subject')
plt.xlabel("Number of films")
plt.ylabel("Subject");
```

## Plot: Number of films by Awards

```python
df['Awards'].value_counts().plot.bar(title='Number of films by awards')
plt.xlabel('Awards?')
plt.ylabel('Number of awards');
```
